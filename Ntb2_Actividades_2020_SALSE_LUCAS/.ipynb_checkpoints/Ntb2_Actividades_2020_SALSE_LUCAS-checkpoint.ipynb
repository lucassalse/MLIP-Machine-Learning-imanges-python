## Ejercicio 2: Carga y Edición de Imagen

a) Cargue una imagen a color utilizando la librería `PIL` y muestre la misma utilizando `matplotlib` (recuerde que debe importar las librerías previamente).

b) Una vez mostrada la imagen, indique cuál es el tamaño, el modo y el formato de la misma.

c) Transformar la imagen de color a escala de grises y mostrar la misma utilizando `matplotlib`.

```python
from PIL import Image
import numpy as np
from matplotlib.pyplot import imshow


fig, axs = plt.subplots(nrows=1, ncols=3, sharex=False)
#Punto a)
I=Image.open("violetas.jpg")  
array_I = np.array(I) #Imagen pasada a matriz utilizando numpy.

#Púnto b)
print("Para la imagen el tamaño es ",I.size,", el modo es ",I.mode," y el formato es", I.format)

#Punto c)
monocroma = np.mean(array_I,axis = 2) #intensidad monocroma=(rojo+verde+azul)/3

#para mostrar gráfico de paleta
l=np.linspace(0,array_I.max(),256)#vector de intensidades
l=l.astype('uint8')

tr=np.zeros((256,40),'uint8')
for i in range(40):
    tr[:,i]=l


fig.set_figheight(8)
fig.set_figwidth(20)
axs[0].imshow(array_I)
axs[0].set_title("Imagen original RGB")
axs[1].imshow(monocroma, cmap = "gray") 
axs[1].set_title("Escala de grises")
axs[2].imshow(tr, cmap = "gray") 
axs[2].set_title("Escala de intensidad")
plt.show()


```

## Ejercicio 3: Formatos de Imagen

a) Cargue una imagen a color utilizando la librería `PIL` (recuerde que debe importarla previamente).

b) Obtenga tres imágenes arreglos utilizando la librería `numpy` : las componentes colorada, verde y azul (RGB) de la misma imagen. 

c) Grafique los tres arreglos o componentes por separado utilizando la librería `matplotlib` (recuerde que debe importarla previamente). Imprima el tamaño de cada una de las componentes utilizando el método `shape`.

d) ¿Qué puede decir al respecto del tamaño de cada una de las componentes? Explique por qué ocurre dicha coincidencia. 

```python
from PIL import Image
import numpy as np
from matplotlib.pyplot import imshow

#Punto a)
I = Image.open("violetas.jpg") 
plt.imshow(I)
array_I = np.array(I)
plt.title('Imagen original')
plt.show()
#Punto b)
rojo  = array_I[:, :, 0] # Obtenemos las matrices completas de cada uno de los colores (0, 1, 2)
verde = array_I[:, :, 1]
azul  = array_I[:, :, 2]

#Punto c)
paleta = "gray"

print("dimensiones de la componente roja (monocroma) =" , rojo.shape)
plt.imshow(rojo, paleta)
plt.title('componente rojo:arreglo MxN')
plt.colorbar()
plt.show()

print("dimensiones de la componente verde (monocroma) =" , verde.shape)
plt.imshow(verde, paleta)
plt.title('componente verde:arreglo MxN')
plt.colorbar()
plt.show()

print("dimensiones de la componente azul (monocroma) =" , azul.shape)
plt.imshow(azul, paleta)
plt.title('componente azul:arreglo MxN')
plt.colorbar()
plt.show()

```

Respuesta pregunta d): Cada una de las componentes de los tres colres **(RGB)**  tienen la misma dimensión que
la imágen original, lo cual tiene sentido ya que cada una de ellas aportan un cierto nivel de intensidad en su
correspondiente color para formar la imagen original.

```python
#Coloque este ejercicio ya que muetra otra manera de realizar cambios en la matriz, mediante concatenaciones
#A diferencia del ejercicio 1 de esta activdad en donde se emplea directamente operar sobre los rangos de la matriz

negro  = np.zeros ((8,8))
blanco = np.ones  ((8,8))

fila1 = np.concatenate((negro, blanco, negro, blanco, negro), axis = 1)
fila2 = 1 - fila1
tablero = np.concatenate((fila1, fila2, fila1, fila2, fila1), axis = 0)

plt.imshow(tablero, cmap='gray')
plt.show()
print (tablero, tablero.shape)
```
