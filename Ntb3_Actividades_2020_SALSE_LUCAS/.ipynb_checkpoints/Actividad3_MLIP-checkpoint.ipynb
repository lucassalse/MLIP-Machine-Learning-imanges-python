### Mejora de Imagen e Histograma

Cargar en la notebook utilizando la librería OpenCV una imagen de manera monocroma y luego:

* Oscurecer la imagen y almacenarla en otra nueva variable.
* Aclarar la imagen y almacenarla en otra nueva variable.
* Aumentar el contraste de la imagen utilizando la ecualización del histograma.
* Mostrar los 4 histogramas de la imagen y aplicar brevemente la naturaleza de cada uno.

```python
### UTILIZAR TANTAS CELDAS DE CÓDIGO COMO SEA NECESARIO
import cv2
import numpy as np
import matplotlib.pyplot as plt


def ajuste_gamma(imagen,gamma):
    #función que dada una imagen (arreglo de tipo uint8), y un valor positivo gamma
    #devuelve otra del mismo tipo pero con ajute gamma, si gamma>1, más oscura y si gamma <1 más clara
    imagenresultado=255*((imagen/255)**gamma)
    imagenresultado=imagenresultado.astype('uint8')
    return imagenresultado

#Almacenamos la imágen de forma monocroma en la variable <img>
img = cv2.imread('violetas.jpg',0) 


#Oscurecer la imagen y almacenarla en otra nueva variable (img_osc)
img_osc = img/2 # T(img)=img/2 Oscurece la imagen

print("Intensidad Píxel (0,0) original: ",img[0,0],"Su tipo de dato es", type(img[0,0]))
print("Intensidad Píxel (0,0) luego del oscurecimiento: ",img_osc[0,0],"\nTipo de dato ", type(img_osc[0,0]),
       "\nLa nueva matriz imagen es de tipo:", type(img_osc), "\n")

# Ahora debemos cambiar el tipo de datos a uint8
img_osc = img_osc.astype('uint8') 

print("Después de la conversión de tipo, la imagen oscurecida tiene un tipo de dato ",type(img_osc), 
      "\nY cada elemento es de tipo ",type(img_osc[0,0]), " luego de la conversion que hemos realizado") 
print()
print('Máximo de la imagen original= ',img.max())
print()
print('Máximo de la imagen oscurecida= ',img_osc.max())

#------------------Aclarar la imagen y almacenarla en otra nueva variable---------------------

img_acl = ajuste_gamma(img,0.2)

print("Intensidad Píxel (0,0) original: ",img[0,0],"Su tipo de dato es", type(img[0,0]))
print("Intensidad Píxel (0,0) luego del aclaramiento: ",img_acl[0,0],"\nTipo de dato ", type(img_acl[0,0]),
       "\nLa nueva matriz imagen es de tipo:", type(img_acl), "\n")

print('Máximo de la imagen original= ',img.max())
print()
print('Máximo de la imagen alcarada= ',img_acl.max())


# Aumentar el contraste de la imagen utilizando la ecualización del histograma.
img_equalizada = cv2.equalizeHist(img)



# -------------------------Visualizaciones---------------------------------------------------
fig, axs = plt.subplots(nrows=1, ncols=4, sharex=True)
fig.set_figheight(4)
fig.set_figwidth(15)
axs[0].imshow(img, cmap='gray',vmin=0 ,vmax=255)
axs[0].set_title('Imagen original')

axs[1].imshow(img_osc, cmap='gray', vmin = 0, vmax = 255)
axs[1].set_title('Imagen oscurecida')

axs[2].imshow(img_acl, cmap='gray', vmin = 0, vmax = 255)
axs[2].set_title('Imagen aclarada')

axs[3].imshow(img_equalizada, cmap='gray', vmin = 0, vmax = 255)
axs[3].set_title('Imagen ecualizada')


plt.show()

fig, axs = plt.subplots(nrows=1, ncols=4, sharex=True)
fig.set_figheight(4)
fig.set_figwidth(15)
axs[0].hist(img[:,:].ravel(), 256,[0,256] ) 
axs[0].set_title('Histograma de iamgen original')
axs[0].set_xlabel('intensidad de iluminacion')
axs[0].set_ylabel('cantidad de pixeles')
axs[1].hist(img_osc[:,:].ravel(), 256,[0,256] ) 
axs[1].set_title('Histograma de imagen oscurecida')
axs[1].set_xlabel('intensidad de iluminacion')
axs[1].set_ylabel('cantidad de pixeles')

axs[2].hist(img_acl[:,:].ravel(), 256,[0,256] ) 
axs[2].set_title('Histograma de imagen aclarada')
axs[2].set_xlabel('intensidad de iluminacion')
axs[2].set_ylabel('cantidad de pixeles')

axs[3].hist(img_equalizada[:,:].ravel(), 256,[0,256] ) 
axs[3].set_title('Histograma de imagen ecualizada')
axs[3].set_xlabel('intensidad de iluminacion')
axs[3].set_ylabel('cantidad de pixeles') 


plt.show()
```

#### Explicación histogramas
Al oscurcer una imagen, lo que se hace es disminuir el valor de cada uno de los pixeles, por esta razón al tener un valor menor debemos ver que el histograma se encuentra desplazado hacia la izquierda.

Al aclarar una iamgen, cada uno de los pixeles aumentan su valor por lo tanto el histograma se ve desplazado hacia la derecha, dado que tenemos valores más elevados en cada píxel.

En el último histograma (visto desde izquierda a derecha), podemos ver que la distribución de los valores de los pixeles es de forma uniforme por la aplciación de la ecualización en la imagen original, generando una imágen más "homogenea" en ese sentido.





### Transformaciones

Cargue una imagen de a elección y realice lo siguiente:

* Cambie la escala de la imagen cargada por una que sea mayor, y luego por una que sea menor. Almacene las imágenes en nuevas variables, muéstrelas e imprima sus dimensiones con el atributo ".shape".
* Rote la imagen utilizando los ángulos 30, 45 y 90. Almacene y muestre las nuevas imágenes rotadas. ¿Sufrieron algún cambio los píxeles de alguna de las imágenes? ¿Por qué?

```python
### UTILIZAR TANTAS CELDAS DE CÓDIGO COMO SEA NECESARIO
import cv2

# Cambie la escala de la imagen cargada por una que sea mayor,
# y luego por una que sea menor. Almacene las imágenes en nuevas variables, 
# muéstrelas e imprima sus dimensiones con el atributo ".shape".
# ----------------------------------------------------------------------------------
imagen =cv2.imread('violetas.jpg')  # Leemos la imagen a escalar
(h, w) = imagen.shape[:2]           # Almacenamos tanto el largo como el ancho respectivamente
print ("El h es: ", h)
print ("El w es: ", w)
k = 1.5                             # Factor de aumento     en la imagen
d = 2                               # Factor de disminucion en la imagen

imagen_mayor = cv2.resize(imagen,( int(w*k), int(h*k)))     #fijamos un tamaño mayor (920, 1080)
imagen_menor = cv2.resize(imagen,(int(w/d), int(h/d)))      #fijamos un tamaño menor (380, 512)

# Imagen original
text_img0 = "El tamanio de la imagen original es: " + str(imagen.shape)
print(text_img0)
cv2.imshow(text_img0, imagen) # A su vez mostramos el tamanio de la imagen original en el cuado de imagen.
cv2.waitKey(0)

# Imagen mas grande
text_imagen_mayor = "El tamanio de la imagen mayor es: " + str(imagen_mayor.shape)
print(text_imagen_mayor)
cv2.imshow(text_imagen_mayor, imagen_mayor)
cv2.waitKey(0)

# Imagen mas chica
text_imagen_menor = "El tamanio de la imagen menor es: " + str(imagen_menor.shape)
print(text_imagen_menor)
cv2.imshow(text_imagen_menor, imagen_menor)
cv2.waitKey(0)

cv2.destroyAllWindows() #destruye las ventanas creadas

# ------------------------------------------------------------------------------------------------------

# Para rotar la imagen calculamos el centro de la misma
centro = (w / 2, h / 2)
escala = 1

# Calculamos las matrices de rotacion
m_90 = cv2.getRotationMatrix2D(centro, 90, escala)
m_30 = cv2.getRotationMatrix2D(centro, 30, escala)
m_45 = cv2.getRotationMatrix2D(centro, 45, escala)


rotated_img_90 = cv2.warpAffine(imagen, m_90, (w,h))
rotated_img_30 = cv2.warpAffine(imagen, m_30, (w,h))
rotated_img_45 = cv2.warpAffine(imagen, m_45, (w,h))

cv2.imshow('Original Image',imagen)
cv2.waitKey(0) 
cv2.destroyAllWindows()

cv2.imshow("90 grados",rotated_img_90)
cv2.waitKey(0) 

cv2.imshow("30 grados",rotated_img_30)
cv2.waitKey(0) 

cv2.imshow("45 grados",rotated_img_45)
cv2.waitKey(0) 
cv2.destroyAllWindows()

```

#### ¿Sufrieron algún cambio los píxeles de alguna de las imágenes? ¿Por qué?
Cuando se efectuaron las diferentes rotaciones, se peude observar que perdemos información de algunos píxeles, esto se debe a que una vez realizada la rotación las dimensiones pasan a ser otras para la imagen rotada. Es por esto que para solucionarlo se debería ajsutar el tamaño de la salida en cada una de las diferentes rotaciones.

### Filtros

Cargue una imagen a elección y realice lo siguiente:

* Aplíquele un filtro gaussiano utilizando la librería OpenCV y muestre la imagen filtrada. Explique lo que sucede al aplicar dicho filtro.

* Aplíquele un filtro Sobel utilizando la librería OpenCV y muestre la imagen filtrada. Explique brevemente lo que sucede al aplicar dicho filtro. Muestre los bordes horizontales y verticales, por separado, de la imagen aplicándole previamente el valor absoluto. ¿Cuál sería la diferencia si deseo utilizar un filtro de Laplace en lugar de Sobel?

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('violetas.jpg')#carga la imagen BGR como arreglo numpy mxnx3 uint8 
img_original = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)#BGR a monocroma numpy mxn uint8

n = 15

gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)#BGR a monocroma numpy mxn uint8
blur = cv2.GaussianBlur(gris,(n,n),0)

fig, axs = plt.subplots(nrows=1, ncols=2)
fig.set_figheight(8)
fig.set_figwidth(16)

axs[0].set_title('Imagen original')
axs[0].imshow(img_original, cmap='gray', vmin=0, vmax=255)
axs[1].set_title('Imagen con ruido Gaussiano')
axs[1].imshow(blur, cmap='gray', vmin=0,vmax=255)


```

#### Explicación de lo que sucede al aplicar el filtro Gaussiano
El filtro Gaussiano realiza un suavizado de la imágen provocando que ésta se desenfoque y se pierdan algunos de los detalles. Su utilidad es disminuir (no dejar pasar), las altas frecuencias (filtro pasa-bajo), es útil para reducir el ruido que podría tener una imágen. A mayor tamaño kernel mayor desenfoque.


```python


img2_original = cv2.imread('edificio.jpg', cv2.IMREAD_GRAYSCALE)
# Filtro Sobel
sobel_h = cv2.Sobel(img2_original, ddepth = cv2.CV_64F, dx = 1, dy = 0, ksize = 3) # Derivada en x (horizontal)
sobel_v = cv2.Sobel(img2_original, ddepth = cv2.CV_64F, dx = 0, dy = 1, ksize = 3) # Derivada en y (vertical)
sobel_r = np.sqrt(sobel_h ** 2 + sobel_v ** 2) # Combinamos ambos gradientes para calcular la resultante

# Filtro de Laplace
laplacian = cv2.Laplacian(img2_original, cv2.CV_64F)

# Graficamos las 4 imagenes anteriores
fig, axs = plt.subplots(nrows=2, ncols=2)
fig.set_figwidth(8)
fig.set_figheight(10)

axs[0,0].set_title('Filtro Laplaciano')
axs[0,0].imshow(15*laplacian, cmap='gray', vmin=0, vmax=255)
axs[0,1].set_title('Filtro Sobel (resultante)')
axs[0,1].imshow(sobel_r, cmap='gray', vmin=0,vmax=255)
axs[1,0].set_title('Filtro Sobel (horizontal)')
axs[1,0].imshow(sobel_h, cmap='gray', vmin=0,vmax=255)
axs[1,1].set_title('Filtro Sobel (vertical)')
axs[1,1].imshow(sobel_v, cmap='gray', vmin=0,vmax=255)
```

#### Que sucede al aplicar el filtro Sobel y el de Laplace
Cuando se aplica el filtro Sobel lo que se hace es destacar bordes verticales u horizontales. Por lo tanto podemos hacer una separación donde en cada una se distinguen dichos bordes.
La representación resultante será la suma de los valores absolutos de los mismos, que como podemos ver, distingue bastante bien todos los bordes de la imágen original.

Si se desea aplicar el filtro de Laplace en lugar de Sobel, se necesita un sólo kernel para detectar los bordes horizontales y verticales. Hay que tener en cuenta que éste tipo de filtro es demasiado sensible al ruido, por lo que en general primero se utiliza un suavizado Gaussiano.
